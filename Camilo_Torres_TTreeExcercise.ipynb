```
import ROOT
```

```
from array import array
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para un par de dados' )
```

```
x = array( 'f', [ 0 ] )
y = array( 'f', [ 0 ] )
z = array( 'f', [ 0 ] )
```

```
MyTree.Branch( 'X', x, 'x/F' )
MyTree.Branch( 'Y', y, 'y/F' )
MyTree.Branch( 'Z', z, 'z/F' )
```

```
from random import gauss
from random import uniform
```

```
mu1, sigma1 = 6.0, 1 # mean and standard deviation
mu2, sigma2 = 5.0, 1 # mean and standard deviation

for i in range(10000):
  x[0] = gauss(mu1, sigma1)
  y[0] = gauss(mu2, sigma2)
  z[0] = uniform(0.0,10.0)
  
  MyTree.Fill()
```

```
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("x*y>>h1(20,0.0,70)") #Bines, dato inicial, dato final
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*z>>h2(14,0.0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("x*y*z>>h3(20,0.0,700)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
#No tiene sentido alguno poner la condición Z==5, debido a que los números son float entre (0.0,10.0) y estos muy dificilmente son un entero que sea 5
#por ello lo que hago es poner la condición de restar z-5 y si este número es menor que un epsilon (0.01) se acepta como válido 
MyTree.Draw("x>>h4(10,0.0,10)","z-5<=0.01")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
#Similarmente al ejercicio anterior
#No tiene sentido alguno poner la condición y==5, debido a que los números son float con una distribución gaussiana y estos muy dificilmente son un entero que sea 5
#por ello lo que hago es poner la condición de restar y-5 y si este número es menor que un epsilon (0.01) se acepta como válido 
MyTree.Draw("x:z>>h5(20,0.0,10,20,0.0,10)","y-5<=0.01","lego")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
