```python
from numpy import sqrt, pi, exp, linspace, save, random, load
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x=load("x_spectra (4).npy")
y=load("y_spectra (1).npy")
```

```python
plt.scatter(x, y, color="m")
plt.show()
```

```python
def recta_gaussian(x,a,b,amp1,cen1,wid1,amp,cen,wid):
    return a*x+b+amp1*exp(-(x-cen1)**2/wid1)+ amp*exp(-(x-cen)**2/wid)
```

```python
init_vals = [-0.05,100,120,10,1,130,30,2]
# for [amp, cen, wid]
best_vals, covar = curve_fit(recta_gaussian, x, y, p0=init_vals)

print (best_vals)

plt.scatter(x, y, color="b")
plt.plot(linspace(0,100,1000),recta_gaussian(linspace(0,100,1000),best_vals[0],best_vals[1],best_vals[2],best_vals[3],best_vals[4],best_vals[5],best_vals[6],best_vals[7]),"r-")

plt.show()
```
